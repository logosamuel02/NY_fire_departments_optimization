  ## Scoring

- Shape Area y Population -> densidad de población. + aumenta score.
- Nº estaciones más cercanas -> establecer threshold manualmente. + disminuye score
- Capacidad estación. + disminuye score
- Distancia al vecindario -> los 5 vecindarios más cercanos a la estación(voronoi). + disminuye score
- Incidencias
  - Nº de incidencias por vecindario. + aumenta score
- Dia y noche.
  - Turno de noche --> aumenta score ligeramente, salvo en unos casos
  - Para turno de noche, fijar una disminución del score para Manhattan y fijar un aumento para Brooklyn y Queens:
    $S_{incidencias-Brooklyn-Night} = 3 * S_{incidencias-Brooklyn-Day}$
    $S_{incidencias-Queens-Night} = 2 * S_{incidencias-Queens-Day}$
    $1.8 * S_{incidencias-Manhattan-Night} = S_{incidencias-Manhattan-Day}$



$S_{Final} = S_{population} + S_{distancias} + S_{incidencias}$
